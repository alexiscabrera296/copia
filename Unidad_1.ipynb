# <left><i><font color="3380ff"> Conjunto </font></i></left>

Es una Colección de objetos en un Universo (Contexto) $\mathbb{U}$  los cuales cumplen o hacen verdadera una proposición lógica abierta $\mathbb{P}$, es decir :
$$A= \{\circledast\in\mathbb{U}\mid P(\circledast) \}$$

$\color{orange}{\text{Conjunción:}}$ Una proposición lógica T es una conjunción si existen dos proposiciones lógicas P, Q, tal que P y Q se puedan expresar como "P y Q" y se denota por  $\mathbf{P \land Q}$

$\color{orange}{\text{Disyunción:}}$ Una proposición lógica T es una disyunción si existen dos proposiciones lógicas P, Q, tal que se puedan expresar como "P o Q" y se denota por $\mathbf{P \lor Q}$

$\color{orange}{\text{Negación:}}$ Una proposición lógica T es una negación siempre que se pueda expresar como "Es falso que P","No P",..., y se denota por $\mathbf{\neg P}$

$\color{orange}{\text{Condicional / Implicación:}}$ Una proposición lógica T para la cual existen dos proposiciones P y Q tal que P y Q se puedan expresar como''Si P entonces Q ",''Dado P entonces Q"
se llamara Implicación o Condicional y se denota por $\mathbf{P \Rightarrow Q}$

$\color{springgreen}{\textbf{NOTA: }} \color{springgreen}{\text{No denotar la implicación como }} \color{springgreen}{\rightarrow ,} \color{springgreen}{\text{ se debe denotar por }} \color{springgreen}{\Rightarrow}$

$\color{orange}{\text{Bicondicional:}}$ Una proposición lógica T se puede expresar como ``P si solo si Q '' , ''P entonces Q y Q entonces P'' se llamara bicondicional y se denota por $\mathbf{P \Longleftrightarrow Q}$

$\color{orange}{\text{Tautologia:}}$ Es una proposición lógica que no depende del contexto para ser cierta.

## <left><i><font color="3380ff"> Definicion de valores de verdad </font></i></left>

$\color{orange}{\text{Conjunción:}}$ Una conjuncion es verdadera siempre que P y Q sean ciertas, en otros casos es falsa 

$\color{orange}{\text{Disyunción:}}$ Una disyunción es falsa simpre que P es falsa y Q es falsa, en otros casos sera verdadera lógica T es una disyunción
si existen dos proposiciones lógicas P, Q, tal que se puedan expresar como "P o Q" y se denota por $\mathbf{P \lor Q}$


$\color{orange}{\text{Negación:}}$ Una negación es verdadera siempre que P es falsa

$\color{orange}{\text{Condicional / Implicación:}}$ Una implicación es falsa siempre que P es verdadero y Q es falsa en otros casos es verdadera

$\color{orange}{\text{Bicondicional:}}$ Una bicondicional es falsa siempre que P es falso y Q es falso, o bien P es verdadero y Q es falso; en otros casos es verdadero.

## <left><i><font color="3380ff"> Definicion </font></i></left>

Sean A y B conjuntos en el universo $\mathbb{U}$ con $A=\bullet\in\mathbb{U}:P(\bullet)$ y $B=\ast\in\mathbb{U}:P(\ast)$

Asi definimos:
1. El conjunto $\color{cyan}{\text{Union}}$ de A y B, como el conjunto:
$$A\cup B:=\{\Box\in\mathbb{U}:P(\Box)\lor
Q(\Box)\}$$
2. El conjunto $\color{cyan}{\text{Insterseccion}}$ de A con B, como el conjunto:
$$A\cap B:=\{\oplus\in\mathbb{U}:P(\oplus)\lor
Q(\oplus)\}$$
3. El conjunto $\color{cyan}{\text{Complemento}}$ de A, como el conjunto:
$$\mathbb{U}-A=A^c:=\{\bigstar\in\mathbb{U}:\neg P(\bigstar)\}$$
4. El conjunto $\color{cyan}{\text{Diferencia}}$ de $A$ con $B$, como el conjunto:
$$A-B = A \backslash B:= \{\clubsuit \in \mathbb{U} : P(\clubsuit) \land \neg Q(\clubsuit)\}$$
5. El conjunto A es $\color{cyan}{\text{Subconjunto}}$ de B sii:
$$\forall \ \$\in\mathbb{U}:=P(\$)\Leftarrow Q(\$)\text{ y se denota por } A\subset B$$

$\color{springgreen}{\textbf{NOTA: }} \color{springgreen}{\text{No impide que A y B sean iguales}}.$

6. El conjunto A es $ \color{cyan}{\text{Subconjunto propio}}$ de B sii:

$$
A\subset B \land \left[ \ \exists\ \bullet \in\mathbb{U}:Q(\bullet) \land \neg P  (\bullet)\  \right] \text{ y se denota por } A\subseteq B
 $$

$\color{springgreen}{\textbf{NOTA: }} \color{springgreen}{\text{Esta vez A y B no pueden ser iguales}}.$

7. El conjunto A es $\color{cyan}{\text{igual}}$ a B sii:
$$\forall \ \star \in\mathbb{U}:=P(\star)\Longleftrightarrow Q(\star)\text{ y se denota por } A = B$$

$\color{magenta}{\text{COROLARIO }} A=B\Longleftrightarrow A\subset B\land B\subset A$

El vacio es el conjunto cuya proposicion logica es una contradiccion y se denota por $\empty$

# <left><i><font color="3380ff"> Funciones </font></i></left>

Para dos objetos a y b, defininamos a la pareja ordenada a con b, en ese orden como 
$$(a,b):=\{a,\{b\}\}\hspace{1cm}(b,a):=\{b,\{a\}\}$$

$\textbf{Terna}$

$$(a,b,c):=\{a,\{b,\{c\}\}\}\hspace{1cm}(a,c,b):=\{a,\{c,\{b\}\}\}$$

Dados dos conjuntos A y B $\color{red}{\text{(No nescesariamente en el mismo universo)}}$, al conjunto de parejas ordenadas (a,b) con $a\in A$ y $b\in B$, se le llama $\color{turquoise}{\text{Producto Cartesiano}}$ de A con B en ese orden. Dicho conjunto se denota por :
$$\color{red}{A\times B:=\{(a,b)\mid a\in A \land b\in B \}}$$

Un conjunto $F\subset A\times B$, con A y B conjuntos, cual cumple:
$$
1. \forall a \in A, \exists b\in B (a,b)\in F \equiv \forall a \in A, \exists ! b \in B (a,b) \in F \\
2. (a,b),(a,c)\in F \Rightarrow b = c \equiv \forall a \in A, \exists ! b \in B (a,b) \in F
$$

Aqui el conjunto A se denomina dominio de $f$ y se denota $Dom f$. La proposicion logica abierta que define a $f$ se llama regla de correspondencia.

Si tenemos $(a,b)\in F$, entonces b se llama imagen de a y se denota $f(a)=b$

El conjunto B se llama codominio o contradominio de $f$ y se denota $cod(f)$

$\color{springgreen}{\text{NOTA: El rango es la dimension del conjunto imagen de una transformacion lineal, no confundir con el siguiente conjunto.}}$

Al conjunto de todas las imagenes de f  la denotaremos como $Img(f):=\{f(a)\mid a\in A \}$

Otra manera de definirlo es $f(a):=\{f(a)\mid a\in A \}$.

Si $C\subset A$, el conjunto 
$$f(C):=\{f(a)\in B \mid a\in C \}\text{ es la imagen de c}$$
Si $D\subset B$, el conjunto 
$$f^{-1}(D):=\{ a\in A \mid f(a)\in D \}\text{ se le llama pre-imagen, imagen inversa de D}$$

$\color{springgreen}{\text{NOTA: No es la funcion inversa}}$

Si $D=\{d\}$, $f^{-1}(d)=f^{-1}(D)$

La grafica de una funcion 
$$Graf(f)=\{(a,f(a)\mid a\in A \}$$

$\color{springgreen}{\text{NOTA: No confundir la grafica con la representacion esquematica de la grafica}}$

La funcion con $DomA$, $CodB$ y regla de correspondencia $f(A)=B$ de denota como 

$$
\begin{array}{ccc}
f:A \rightarrow{B} & f:A \rightarrow{B}  & A \rightarrow{B} \\
a \mapsto b & f(a)=b & a \mapsto b
\end{array}
$$

Ejemplo:
1. $ f:\mathbb{R}\to\mathbb{R},f(x)=2$

$\color{springgreen}{\text{NOTA: Esta funcion se llama funcion constante}}$

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Definir la función constante
def funcion_constante(c=2.0):
    x = np.linspace(-5, 5, 500)
    y = np.full_like(x, c)  # Crea un array donde todos los valores son `c`
    
    # Configurar el gráfico
    plt.figure(figsize=(8, 4))
    plt.plot(x, y, label=f'$f(x) = {c}$', color='blue', linewidth=2)
    plt.axhline(y=c, linestyle='--', alpha=0.3, color='blue')
    plt.title('Función Constante $f(x) = c$', fontsize=14)
    plt.xlabel('$x$', fontsize=12)
    plt.ylabel('$f(x)$', fontsize=12)
    plt.legend(loc='upper right')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.ylim(-1, 5)  # Rango fijo en el eje Y para mejor visualización
    plt.show()

# Crear el widget interactivo
interact(
    funcion_constante,
    c=FloatSlider(
        value=2.0,
        min=-3.0,
        max=5.0,
        step=0.5,
        description='Valor de $c$:',
        continuous_update=True  # Actualiza en tiempo real al mover el slider
    )
);
```
